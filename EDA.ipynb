```python
#perform EDA for Energy/energydata.csv
import pandas as pd
import numpy as np
df = pd.read_csv('data/energydata.csv')
```

```python
#check for negativ
df.describe()
```

```python
ac_power = 1.5
tv_power = 0.1
df['AC_hours'] = df['amount_paid'] / (ac_power * df['ac'] * 30.4)
df['TV_hours'] = df['amount_paid'] / (tv_power * df['tv'] * 30.4)
```

```python
#save df as csv
```

```python
#change NaN and inf to 0
df = df.replace([np.inf, -np.inf], 0)
```

```python
df.describe()
```

```python
df.to_csv('Energy/energydata.csv', index=False)
```
